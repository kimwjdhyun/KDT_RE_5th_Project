```python
folder = Path("./강원도_구역별_강수량(2020-2025)")
output_excel = "강원도_월별강수량_통합.xlsx"

with pd.ExcelWriter(output_excel, engine="openpyxl") as writer:
    for csv_file in folder.glob("*.csv"):
        print(f"처리 중: {csv_file.name}")

        df = pd.read_csv(
            csv_file,
            encoding="cp949",
            engine="python",
            on_bad_lines="skip",
            skiprows=8
        )
        sheet_name = csv_file.stem[:31]

        df.to_excel(writer, sheet_name=sheet_name, index=False)
```
