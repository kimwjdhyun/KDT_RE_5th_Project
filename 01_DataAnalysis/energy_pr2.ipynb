```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import preprocessing as pre


df1 = pd.read_excel(
    './data/processed/강원도_월별기온_통합.xlsx', 
    sheet_name=None
)

processed_dfs = {}

use_cols = [
    '지점번호',
    '지점명',
    '일시',
    '평균기온(℃)',
    '평균최고기온(℃)',
    '최고기온(℃)',
    '평균최저기온(℃)',
    '최저기온(℃)',
]

for sheet, df in df1.items():
    #컬럼명 정리
    df.columns = (
        df.columns
          .str.strip()
          .str.replace(" ", "")
    )

    # 필요한 컬럼만 선택 (이게 df1_use 역할)
    df_use = df[[c for c in use_cols if c in df.columns]].copy()
    
    # 결측치 처리
    for col in ["평균기온(℃)", "평균최고기온(℃)", "평균최저기온(℃)"]:
        if col in df_use.columns:
            df_use = pre.handle_missing(df_use, col)

    processed_dfs[sheet] = df_use

processed_dfs["강원도(강릉)"]
```
