```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time

options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(options=options)

BASE_URL = "https://k-re100.or.kr/bbs/board.php?bo_table=sub2_2_1&page={}"
TOTAL_PAGES = 1

results = []

for page in range(1, TOTAL_PAGES + 1):
    driver.get(BASE_URL.format(page))
    time.sleep(2)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    rows = soup.select("table tbody tr")

    for row in rows:
        title_tag = row.select_one("a")

        # ğŸ”‘ ê²Œì‹œê¸€ì´ ì•„ë‹Œ í–‰ ì œì™¸
        if title_tag is None:
            continue

        title = title_tag.get_text(strip=True)
        link = title_tag.get("href")

        # ë‚ ì§œëŠ” ë§ˆì§€ë§‰ td
        tds = row.select("td")
        date = tds[-1].get_text(strip=True) if tds else None

        if link and link.startswith("/"):
            link = "https://k-re100.or.kr" + link

        results.append({
            "ì œëª©": title,
            "ì‘ì„±ì¼": date,
            "ë§í¬": link
        })

    print(f"{page} í˜ì´ì§€ ìˆ˜ì§‘ ì™„ë£Œ")

driver.quit()

df = pd.DataFrame(results)
df.to_csv("k_re100_board.csv", index=False, encoding="utf-8-sig")

print("í¬ë¡¤ë§ ì™„ë£Œ: k_re100_board.csv ìƒì„±")
```
