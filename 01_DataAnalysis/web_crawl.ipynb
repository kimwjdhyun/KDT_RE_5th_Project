```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 강원특별자치도 포털 정책 자료 페이지 (예시 URL)
url = "https://state.gwd.go.kr/portal/bbs/list.do?bbsId=BBSMSTR_000000000001"  

response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")

# 공지사항/정책자료 제목과 링크 추출
rows = soup.select(".board_list tbody tr")

data = []
for row in rows:
    title = row.select_one(".title a").get_text(strip=True)
    link = row.select_one(".title a")["href"]
    date = row.select_one(".date").get_text(strip=True)
    data.append({"제목": title, "링크": link, "날짜": date})

df = pd.DataFrame(data)
print(df.head())

# CSV 저장
# df.to_csv("policy_data.csv", index=False, encoding="utf-8-sig")
```
