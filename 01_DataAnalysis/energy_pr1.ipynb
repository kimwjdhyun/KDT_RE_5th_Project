```python
#행열 정리 함수

def clean_structure(df):
    df = df.iloc[:-3, :-3]      # 뒤쪽 행/열 제거만
    return df.reset_index(drop=True)

clean_structure(df1)

df1 = clean_structure(df1)
```

```python
#숫자로 변환 못할 경우 그 값을 NaN으로 대체 코드
pd.to_numeric(
    df1["강수량(mm)"], 
    errors="coerce"
)

pre = df1["강수량(mm)"].mean().round(1)
#결측값 평균 값으로 대체
df1["강수량(mm)"] = df1["강수량(mm)"].fillna(pre)
df1
```

```python
#결측값 처리 (평균 값으로 대체)
def handle_rainfall_missing(df,col):
    if not pd.api.types.is_numeric_dtype(df[col]):
        df[col] = pd.to_numeric(df[col], errors="coerce")
    
    pre = df[col].mean().round(1)
    df[col] = df[col].fillna(pre)

    return df

df = clean_structure(df1)
df = handle_rainfall_missing(df,"강수량(mm)")

df
```

```python
#함수 적용해 전체 데이터 전처리 적용

df1 = pd.read_excel(
    './data/processed/강원도_월별강수량_통합.xlsx', 
    sheet_name=None
)

processed_dfs = {}

for sheet, df in df1.items():
    df = clean_structure(df)
    df.columns = df.columns.str.strip()

    df = handle_rainfall_missing(df,"강수량(mm)")
    df = handle_rainfall_missing(df,"일최다강수량(mm)")
    
    processed_dfs[sheet] = df

processed_dfs["강원도(강릉)"]
```

```python
# 여러번 전처리 처리하면서 개선한 코드로 적용

import preprocessing as pre
#만들어 뒀던 결측치 처리 py파일 불러움.

processed_dfs = {}
#빈 딕셔너리 만드는 건 시트별로 전처리된 DataFrame을 저장하기 위해 사용.

use_cols = [
    '지점번호',
    '지점명',
    '일시',
    '강수량(mm)',
    '일최다강수량(mm)'
]

print("총 시트 수:", len(df1))

for sheet, df in df1.items():
    print("처리 중:", sheet) #엑셀 시트가 처리될 때마다 출력

    df.columns = df.columns.str.strip().str.replace(" ", "") #컬럼 명 정리
    #특정 컬럼만 추려내서 데이터화 -원본 dfs 살리기 위해 copy 씀.
    df_use = df[[c for c in use_cols if c in df.columns]].copy() #컬럼 
    processed_dfs[sheet] = df_use
    #전처리 된 df_use를 sheet라는 키로 저장해라.

    # 결측치 처리
    for col in ["강수량(mm)","일최다강수량(mm)"]:
            if col in df_use.columns:
                df_use = pre.handle_missing(df_use, col)
    processed_dfs[sheet] = df_use

processed_dfs["강원도(태백)"]
```

```python
#결측치가 없는지 확인
len(processed_dfs["강원도(강릉)"]) - processed_dfs["강원도(강릉)"].count()

#확인 도중 최다 강수량에 결측있으면 최다강수량일자도 처리되어 있고 이는 
#평균으로도, 임의로도 처리할 수 없는 걸 파악해.
#일자 칼럼을 쓰지 않기로 정함. 
```

```python
#전처리 과정 거친 데이터를 엑셀 시트로 저장
#추후 시각화 과정에서 정제된 데이터를 바로 쓰기 위함.

output_path = "./data/processed/New_강원도_월별강수량.xlsx"

with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    for sheet, df in processed_dfs.items():
        df.to_excel(
            writer,
            sheet_name=sheet,
            index=False
        )
```
